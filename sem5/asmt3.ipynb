Implement Python Libraries for ML application such as Pandas,Matplotlib and Seaborn
(Using Marksheet Dataset)

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris()

iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df['species'] = pd.Categorical.from_codes(iris['target'], iris['target_names'])

print(iris_df.head())
```

```python
print(iris_df.isnull().sum())
```

```python
filtered_data = iris_df[iris_df['sepal length (cm)'] > 5]
print(filtered_data.head())
```

```python
grouped_mean = iris_df.groupby('species').mean()
print(grouped_mean)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(iris_df.index, iris_df['sepal length (cm)'], label='Sepal Length (cm)')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length Line Plot')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(iris_df['sepal length (cm)'], bins=20, color='skyblue')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c='green')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width Scatter Plot')
plt.show()
```

```python
sns.pairplot(iris_df, hue='species')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr_matrix = iris_df.iloc[:, :-1].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.violinplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title('Violin Plot of Sepal Length for Each Species')
plt.show()
```

```python
sns.lmplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df)
plt.title('Sepal Length vs Petal Length Regression Plot')
plt.show()
```
